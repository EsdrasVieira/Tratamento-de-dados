```python
import numpy as np
import matplotlib.pyplot as plt

# Dados
plataformas = ['Namorado 2', 'P 26', 'P 32', 'P IX', 'P XXXIII']
estacoes = ['Inverno', 'Outono', 'Primavera', 'Vero']
forca_media = np.array([
    [15.4, 12.4, 16.5, 16.1],
    [18.8, 11.9, 17.13, 20.6],
    [15.0, 12.0, 16.9, 20.4],
    [15.5, 12.5, 16.0, 19.4],
    [14.8, 11.8, 17.1, 20.4]
])

# Paleta de cores
cores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

# Criando o grfico
plt.figure(figsize=(10, 6))

# Criando o grfico de linhas
for i in range(len(plataformas)):
    plt.plot(estacoes, forca_media[i], marker='o', label=plataformas[i], color=cores[i])

# Adicionando ttulo e rtulos nos eixos
plt.xlabel('Estao do Ano')
plt.ylabel('Fora Mdia (KN)')
plt.title('Perfil de Fora ao Longo do Ano por Plataforma e Estao do Ano')

# Adicionando legenda no canto inferior direito
plt.legend(title='Plataforma', loc='lower right')

# Adicionando grades
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Ajustando o layout
plt.tight_layout()

# Exibindo o grfico
plt.show()
```
